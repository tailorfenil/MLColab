# Introduction to Gradient Descent

Gradient descent is an optimization algorithm used to minimize the cost function of a model. It is commonly used in machine learning to update the model parameters iteratively until the optimal solution is found. The idea behind gradient descent is to calculate the gradient of the cost function with respect to the parameters and update them in the opposite direction of the gradient until the minimum is reached.

The gradient descent algorithm can be described as follows:

1. Initialize the model parameters with random values.
2. Calculate the cost function using the current parameters.
3. Calculate the gradient of the cost function with respect to each parameter.
4. Update each parameter by subtracting the product of the learning rate and the gradient from the current parameter.
5. Repeat steps 2-4 until the cost function converges to a minimum.

The learning rate is a hyperparameter that controls the step size at each iteration. A high learning rate can result in convergence to a minimum quickly, but it may overshoot and oscillate around the minimum. On the other hand, a low learning rate can result in a slow convergence to the minimum.

There are different variants of gradient descent, including batch gradient descent, minibatch gradient descent, and stochastic gradient descent. These variants differ in the amount of data used to compute the gradient at each iteration and how the model parameters are updated.

# Gradient Descent with Fixed Learning Rate

Let's start by implementing gradient descent with a fixed learning rate in Python.


We will use the following cost function:

f(x) = (x - 5) ** 2 + 20


Let us the plot the function first!

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Create an array of x values
x = np.linspace(-10, 20, 1000)

# Evaluate f(x) for each value of x
y = f(x)

# Plot f(x) versus x
plt.plot(x, y)

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) versus x')

# Display the plot
plt.show()
```

It appears to be a simple quadratic function with a minimum at x = 5. We will use gradient descent to find the minimum of this function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return (x - 5) ** 2 + 20

# Define the gradient of the cost function
def gradient(x):
    return 2 * (x - 5)

# Initialize the model parameter with a random value
x = np.random.randn()

# Set the learning rate
learning_rate = 0.1

# Set the maximum number of iterations
max_iterations = 100

# Initialize the list to store the cost values
cost_values = []

# Iterate until convergence or max iterations
for i in range(max_iterations):
    # Calculate the cost and gradient at the current parameter
    cost = cost_function(x)
    grad = gradient(x)

    # Update the parameter
    x = x - learning_rate * grad

    # Store the cost value
    cost_values.append(cost)

    # Print the progress
    if i % 10 == 0:
        print("Iteration {}: x = {}, cost = {}".format(i, x, cost))

# Plot the cost values
plt.plot(cost_values)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

In the above code, we first define the cost function and its gradient. We then initialize the model parameter with a random value and set the learning rate and maximum number of iterations. We iterate until max iterations and calculate the cost and gradient at each iteration. We update the parameter by subtracting the product of the learning rate and gradient from the current parameter. We also store the cost value at each iteration and print the progress. Finally, we plot the cost values to visualize the convergence.

Let's observe how x changes after each iteration of Gradient Descent!

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Define the gradient of f(x)
def grad_f(x):
    return 2 * (x - 5)

# Define the gradient descent algorithm
def gradient_descent(f, grad_f, x_init, alpha, num_iterations):
    x = x_init
    x_list = [x]
    for i in range(num_iterations):
        x = x - alpha * grad_f(x)
        x_list.append(x)
    return x_list

# Set the hyperparameters
alpha = 0.1
num_iterations = 5

# Run gradient descent on f(x)
x_init = 15
x_list = gradient_descent(f, grad_f, x_init, alpha, num_iterations)

# Create an array of x values
x = np.linspace(-2, 15, 100)

# Evaluate f(x) for each value of x
y = f(x)

# Plot f(x) versus x
plt.plot(x, y)

# Plot the gradient descent iterations
for i in range(len(x_list) - 1):
    x1 = x_list[i]
    y1 = f(x1)
    x2 = x_list[i + 1]
    y2 = f(x2)
    plt.plot([x1, x2], [y1, y2], 'ro-')
    plt.text(x1, y1 + 0.5, round(y1, 2))

# Label the final cost value
x_final = x_list[-1]
y_final = f(x_final)
plt.text(x_final, y_final + 0.5, round(y_final, 2))

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient descent iterations on f(x)')

# Display the plot
plt.show()
```

# Impact of learning rate (alpha) on convergence

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return (x - 5) ** 2 + 20

# Define the gradient of the cost function
def gradient(x):
    return 2 * (x - 5)

# Set the maximum number of iterations
max_iterations = 100

# Initialize the list to store the cost values for different learning rates
learning_rates = [0.001, 0.01, 0.1, 1]
cost_values = [[] for _ in range(len(learning_rates))]

# Iterate over different learning rates
for j, learning_rate in enumerate(learning_rates):
    # Initialize the model parameter with a random value

    np.random.seed(42)
    x = np.random.randn()

    # Iterate until convergence or max iterations
    for i in range(max_iterations):
        # Calculate the cost and gradient at the current parameter
        cost = cost_function(x)
        grad = gradient(x)

        # Update the parameter
        x = x - learning_rate * grad

        # Store the cost value
        cost_values[j].append(cost)

    # Plot the cost values for this learning rate
    plt.plot(cost_values[j], label="learning rate = {}".format(learning_rate))

# Add legend and labels to the plot
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

# Batch Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random customer locations
num_customers = 50
customer_locations = np.random.randn(num_customers)

# Define the objective function (sum of squared distances to customers)
def objective(x):
    return np.sum((customer_locations - x)**2)

# Define the derivative of the objective function
def derivative(x):
    return -2 * np.sum(customer_locations - x)

# Set the learning rate and number of iterations
learning_rate = 0.001
num_iterations = 10

# Initialize the store location to a random value
store_location = 2

# Store the objective function values and store locations at each iteration
objective_values = []
store_locations = []

# Perform gradient descent
for i in range(num_iterations):
    # Compute the objective function value and store the values
    obj_value = objective(store_location)
    objective_values.append(obj_value)

    # Compute the gradient
    grad = derivative(store_location)

    # Update the store location
    store_location -= learning_rate * grad

    store_locations.append(store_location)

# Print the final store location
print("Final store location:", store_location)

# Plot the customer locations and the store location
for j in range(len(store_locations)):
    plt.scatter(store_locations[j], 0, color="red", label="Store location",s=100)
    plt.scatter(customer_locations, np.zeros(num_customers), label="Customer locations")
    plt.xlabel("Location")
    plt.legend()
    plt.title("Iteration:"+str(j))
    plt.show()

# Plot the objective function values during gradient descent
plt.plot(objective_values)
plt.xlabel("Iteration")
plt.ylabel("Objective function value")
plt.title("Gradient descent on customer locations")
plt.show()
```

# Mini batch gradient descent

We need to update the derivative function to compute the gradient of the objective function with respect to a randomly selected subset of samples (customer locations).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random customer locations
num_customers = 50
customer_locations = np.random.randn(num_customers)

# Define the objective function (sum of squared distances to customers)
def objective(x):
    return np.sum((customer_locations - x)**2)

# Define the derivative of the objective function
def derivative(x, batch_size):
    batch_indices = np.random.choice(len(customer_locations), batch_size)
    #print(batch_indices)
    #print("returning",-2 * np.sum(customer_locations[batch_indices] - x, axis=0))
    return -2 * np.sum(customer_locations[batch_indices] - x, axis=0)

# Set the learning rate, number of iterations, and batch size
learning_rate = 0.1
num_iterations = 15
batch_size = 5

# Initialize the store location to a random value
store_location = np.array([2])

# Store the objective function values and store locations at each iteration
objective_values = []
store_locations = []

# Perform minibatch gradient descent
for i in range(num_iterations):
    # Compute the objective function value and store the values
    obj_value = objective(store_location)
    objective_values.append(obj_value)

    # Compute the gradient
    grad = derivative(store_location, batch_size)
    #print("grad=",grad)
    # Update the store location
    store_location = store_location - learning_rate * grad

    store_locations.append(store_location)

# Print the final store location
print("Final store location:", store_location)

# Plot the customer locations and the store location
for j in range(len(store_locations)):
    plt.scatter(store_locations[j], 0, color="red", label="Store location",s=100)
    plt.scatter(customer_locations, np.zeros(num_customers), label="Customer locations")
    plt.xlabel("Location")
    plt.legend()
    plt.title("Iteration:"+str(j))
    plt.show()

# Plot the objective function values during gradient descent
plt.plot(objective_values)
plt.xlabel("Iteration")
plt.ylabel("Objective function value")
plt.title("Minibatch gradient descent on customer locations")
plt.show()
```

# Stochatic Gradient Descent

In stochastic gradient descent, we randomly choose one data point from the dataset to compute the gradient at each iteration, instead of using the entire dataset or a small batch of data points like in batch or minibatch gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random customer locations
num_customers = 50
customer_locations = np.random.randn(num_customers)

# Define the objective function (sum of squared distances to customers)
def objective(x):
    return np.sum((customer_locations - x)**2)

# Define the derivative of the objective function
def derivative(x):
    random_index = np.random.randint(num_customers)
    return -2 * (customer_locations[random_index] - x)

# Set the learning rate, number of iterations, and batch size
learning_rate = 0.2
num_iterations = 30

# Initialize the store location to a random value
store_location = np.array([2])

# Store the objective function values and store locations at each iteration
objective_values = []
store_locations = []

# Perform stochastic gradient descent
for i in range(num_iterations):
    # Compute the objective function value and store the values
    obj_value = objective(store_location)
    objective_values.append(obj_value)

    # Compute the gradient
    grad = derivative(store_location)

    # Update the store location
    store_location = store_location - learning_rate * grad

    store_locations.append(store_location)

# Print the final store location
print("Final store location:", store_location)

# Plot the customer locations and the store location
for j in range(len(store_locations)):
    plt.scatter(store_locations[j], 0, color="red", label="Store location",s=100)
    plt.scatter(customer_locations, np.zeros(num_customers), label="Customer locations")
    plt.xlabel("Location")
    plt.legend()
    plt.title("Iteration:"+str(j))
    plt.show()

# Plot the objective function values during gradient descent
plt.plot(objective_values)
plt.xlabel("Iteration")
plt.ylabel("Objective function value")
plt.title("Stochastic gradient descent on customer locations")
plt.show()
```

```python
import pandas as pd

def C(vals):
    x, y = vals[0], vals[1]

    return x**2 + y**2

# Gradient of cost function
def dC(vals):
    x, y = vals[0], vals[1]

    return np.array([2*x, 2*y])

# Gradient descent algorithm
def gradient_descent(initial_q, learning_rate, num_iterations):
    q_values = [initial_q]
    cost_values = [C(initial_q)]

    # Perform gradient descent
    for i in range(num_iterations):
        current_q = q_values[-1]
        delta_q = -learning_rate * dC(current_q)
        new_q = current_q + delta_q
        q_values.append(new_q)
        cost_values.append(C(new_q))

    return pd.DataFrame({'q_values': q_values, 'cost_values': cost_values})

# Define parameters
initial_q = [1,1]
learning_rate = 0.05
num_iterations = 100

# Perform gradient descent
df = gradient_descent(initial_q, learning_rate, num_iterations)
```
